```python
# gorq model integration

import os
from langchain.chat_models import init_chat_model

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")

model = init_chat_model("groq:qwen/qwen3-32b")
model
```

### Streaming and Batch
Models can stream their output content while it is being generated

```python
# without using streams

from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    model="gpt-4o-2024-08-06",
)

# Invocation
ai_msg = llm.invoke("Give me a 200 words content on christmas")
ai_msg.content
```

```python
# using streams

for chunk in llm.stream("Give me a 200 word content on christmas"):
    print(chunk.content, end="|", flush=True)
```

### Batch

processing multiple inputs in parallel thereby imporving the performance and reducing cost 

```python
responses = llm.batch(
    [
        "Why do parrots have colorful feathers?",
        "Why do aeroplanes fly?",
        "What is quantum computing",
    ],
    config={
        "max_concurrency": 5,  # limit to 5 parallel calls
    },
)

for response in responses:
    print(response)
```
