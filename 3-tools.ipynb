### Tools

Models can make calls to tools which perform tasks such as fetching data from database, searching the web or running code.

https://docs.langchain.com/oss/python/langchain/tools

```python
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
from langchain.chat_models import init_chat_model

model = init_chat_model("groq:qwen/qwen3-32b")
```

```python
# invoking tool calls

from langchain.tools import tool

@tool
def get_weather(location: str) -> str:
    """Get current weather information"""
    return f"The current weather in {location} is sunny with a temperature of 25°C."

agent = model.bind_tools([get_weather])
response = agent.invoke("What is the weather in Boston?")
print(response)

# print tool calls made by the model
for i in response.tool_calls:
    print(f"Tool: {i['name']}")
    print(f"Args: {i['args']}")
```

### Tool execution loop

```python
# step 1: Model generates tool_calls or arguments
messages = [{"role": "user", "content": "what is the weather in Boston?"}]
ai_msg = agent.invoke(messages)
messages.append(ai_msg)

# step 2:  Execute tools with generated arguments and collect results
for tool_call in ai_msg.tool_calls:
    # Execute the tool with the generated arguments
    tool_result = get_weather.invoke(tool_call)
    messages.append(tool_result)

# step 3: Pass the results back to model for final response
final_response = agent.invoke(messages)
print(final_response.text)
```
