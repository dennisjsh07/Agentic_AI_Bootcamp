### Messages

https://docs.langchain.com/oss/python/langchain/messages

```python
# Normal text prompt 

import os
from langchain.chat_models import init_chat_model
from langchain.messages import HumanMessage, AIMessage, SystemMessage

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")

model = init_chat_model("groq:qwen/qwen3-32b")

system_msg = SystemMessage("You are a helpful coding assistant.")
human_msg = HumanMessage("How do i create rest api's?")

# Use with chat models
messages = [system_msg, human_msg]
response = model.invoke(messages)  # Returns AIMessage
print(response.content)
print(response.response_metadata)
```

```python
# Detailed text prompt

system_msg = SystemMessage(
    """
    You are a senior backend developer who is specialized in creating rest api's. 
    help in creating rest api's as per production standard
    """
)
human_msg = HumanMessage("How do i create rest api's? explain with an example")

# Use with chat models
messages = [system_msg, human_msg]
response = model.invoke(messages)  # Returns AIMessage
print(response.content)
print(response.response_metadata)
```

```python
# combination of all including ai msg

ai_msg = "i'd be very happy to help you with the question"

messages = [
    SystemMessage("You are a helpful assistant"),
    HumanMessage("Hey i need help"),
    ai_msg, # insert as if came from a model
    HumanMessage("Great! what is 2+2?")
]
response = model.invoke(messages)
print(response.content)
print(response.response_metadata)
```

### Tool message
https://docs.langchain.com/oss/python/langchain/messages#tool-message

```python
from langchain.messages import AIMessage, ToolMessage

# After a model makes a tool call
# (Here, we demonstrate manually creating the messages for brevity)
ai_message = AIMessage(
    content=[],
    tool_calls=[
        {"name": "get_weather", "args": {"location": "San Francisco"}, "id": "call_123"}
    ],
)

# Execute tool and create result message
weather_result = "Sunny, 72°F"
tool_message = ToolMessage(
    content=weather_result,
    tool_call_id="call_123"  # Must match the call ID
)

# Continue conversation
messages = [
    HumanMessage("What's the weather in San Francisco?"),
    ai_message,  # Model's tool call
    tool_message,  # Tool execution result
]
response = model.invoke(messages)  # Model processes the result
print(response.content)
print(response.response_metadata)
```
